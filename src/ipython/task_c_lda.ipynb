## Load Preprocessing Data

```python
dataset = 'C'
input_from_file = pre.get_data('train', dataset)
test_set = pre.get_data('test', dataset)
```

## Make LDA Model

```python
lda_model, vectorizer, train_data, all_topics, topic_words_dist, map_topic_id = pre.get_model(input_from_file)
```

## Make Polarity Classifier with SVM

```python
svm_polarity_model = pre.polarity_model(lda_model=lda_model, model='svm', vectorizer=vectorizer,
                                    topic_words_dist=topic_words_dist,
                                    train_data=train_data, multi=True, map_topic_id=map_topic_id)
```

### SVM Prediction

```python
svm_prediction = pre.polarity_test(lda_model, svm_polarity_model, vectorizer, topic_words_dist, test_set, map_topic_id=map_topic_id)
```

```python
m.get_accuracy(svm_prediction)
```

```python
m.avg_recall(svm_prediction)
```

```python
m.macro_average_mae(svm_prediction)
```

```python
m.standard_mae(svm_prediction)
```
